```python
"""
Random Forest

"""

import pandas as pd
from sklearn import model_selection
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

class RandomForest_Classifier():
    def __init__(self):
        self.split = [2,3,4,5,7,10]
        self.trees = [3,5,7,10,20,25]
        self.depth = [3,5,9,12]
    
    def dataset_split(self):
        df = pd.read_csv("cardio_preprocessed.csv")  
        X = df.drop('cardio',axis=1)
        y = df.cardio
        return X,y

    def random_forest_model_change_in_estimators(self,X,y):
        
        acc = {}
        
        for j in self.trees:
            acc_score = []
            for i in self.split:
                kfold = model_selection.KFold(n_splits=i)
                model = RandomForestClassifier(n_estimators=j, criterion='gini')
                results = model_selection.cross_val_score(model, X, y, cv=kfold)
                accuracy_score = results.mean()
                acc_score.append(accuracy_score)
                acc[j] = acc_score
        return acc
    
    def random_forest_model_change_in_depth(self,X,y):
        
        acc = {}
        
        for j in self.depth:
            acc_score = []
            for i in self.split:
                kfold = model_selection.KFold(n_splits=i)
                model = RandomForestClassifier(n_estimators=25, max_depth=j, criterion='gini')
                results = model_selection.cross_val_score(model, X, y, cv=kfold)
                accuracy_score = results.mean()
                acc_score.append(accuracy_score)
                acc[j] = acc_score
        return acc
    
    def graph(self, ra_tree_est):
        plt.figure()
        for x,y in ra_tree_est.items():
            plt.plot(self.split, y)
        plt.title("Random Forest Results \n with vary in Number of trees \n Variations in Accuracy \n with K-Fold Splits", pad=20)
        plt.ylabel("Accuracy", fontsize = "large")
        plt.xlabel('K-Fold Splits', fontsize = "large")
        plt.legend(['3 Trees' ,'5 Trees','7 Trees','10 Trees','20 Trees','25 Trees'], loc='lower right')
        plt.show()
        
        
    def graph2(self, ra_tree_est):
        plt.figure()
        for x,y in ra_tree_est.items():
            plt.plot(self.split, y)
        plt.title("Random Forest Results \n with vary in Maximum Depth \n Variations in Accuracy \n with K-Fold Splits", pad=20)
        plt.ylabel("Accuracy", fontsize = "large")
        plt.xlabel('K-Fold Splits', fontsize = "large")
        plt.legend(['3 Max Depth' ,'5 Max Depth','9 Max Depth','12 Max Depth'], loc='lower right')
        plt.show()
        



es_cla = RandomForest_Classifier()
X,y = es_cla.dataset_split()


ra_tree_est = es_cla.random_forest_model_change_in_estimators(X,y)
print(ra_tree_est)

for x,y in ra_tree_est.items():
    print("Number of trees {0} ----> Accuracy Score = {1}". format (x, np.mean(ra_tree_est[x])) )
es_cla.graph(ra_tree_est)

ma_cla = RandomForest_Classifier()
X,y = ma_cla.dataset_split()
ra_tree_est_depth = ma_cla.random_forest_model_change_in_depth(X,y)
print(ra_tree_est_depth)

for x,y in ra_tree_est_depth.items():
    print("Maximum Depth {0} ----> Accuracy Score = {1}". format (x, np.mean(ra_tree_est_depth[x])) )
ma_cla.graph2(ra_tree_est_depth)
```

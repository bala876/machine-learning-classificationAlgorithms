```python
X = df[['gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active' ]][:].values.reshape(70000, 10)
print(X)
```

```python
Split the dataset y values
```

```python
y = df[['cardio']][:].values.reshape(70000, 1)
print(y)
```

```python
 Implementing Voting Classifier Algorithm
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, VotingClassifier
from xgboost import XGBClassifier

model1 = DecisionTreeClassifier(random_state=0)
model2 = BaggingClassifier(n_estimators=10)
model3 = XGBClassifier(n_estimators = 10)

model = VotingClassifier(estimators=[('lr', model1), ('rf', model2), ('gnb', model3)], voting='hard')
model.fit(X,y)
```

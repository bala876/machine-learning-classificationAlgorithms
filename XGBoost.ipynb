```python
import pandas as pd
from sklearn.ensemble import BaggingClassifier

df = pd.read_csv("cardio_preprocessed.csv")
X = df[['gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active' ]][:].values.reshape(70000, 10)
y = df[['cardio']][:].values.reshape(70000, 1)
model = BaggingClassifier(n_estimators=10, random_state=0)
model.fit(X,y)
new_data = [2,168,62,110,80,1,1,0,0,1]
print(model.predict([new_data]))
```

```python
import pandas as pd
from sklearn.ensemble import BaggingClassifier
from sklearn.metrics import classification_report

df = pd.read_csv("cardio_preprocessed.csv")
X_train = df[['gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active' ]][:56000].values.reshape(56000, 10)
y_train = df[['cardio']][:56000].values.reshape(56000, 1)
X_test = df[['gender', 'height', 'weight', 'ap_hi', 'ap_lo', 'cholesterol', 'gluc', 'smoke', 'alco', 'active' ]][56000:].values.reshape(14000, 10)
y_test = df[['cardio']][56000:].values.reshape(14000, 1)

model = BaggingClassifier(n_estimators=10, random_state=0)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
```
